Look at heart rate data

```python
import pandas as pd
import plotly.express as px
import numpy as np

df_heartrate_seconds = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Bellabeat/heartrate_seconds_merged.csv')
df_hourlySteps = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Bellabeat/hourlySteps_merged.csv')
```

```python
df_hourlySteps['Date'] = pd.to_datetime(df_hourlySteps['ActivityHour']).dt.date
```

#**Steps per day**

```python
df_steps_per_day = df_hourlySteps.groupby(['Id', 'Date'], as_index=False)['StepTotal'].sum()
```

#Get average steps per day by user

```python
df_avg_steps_per_day = df_steps_per_day.groupby(['Id'], as_index=False)['StepTotal'].mean()
```

#Average steps per day

```python
df_avg_steps_per_day['StepTotal'].mean().astype(int)
```

#Heart rate spikes of 126 and above look interesting. Is there a pattern for the days they occur on?

#Get the data for when heart rate is 126 or higher

```python
df_gt_125 = df_heartrate_seconds[df_heartrate_seconds['Value']>125]
```

#Get the day of the week

```python
dayOfWeek = pd.to_datetime(df_gt_125['Time']).dt.dayofweek
```

#Exercise seems to ramp up significantly Thursday through Sunday. Maybe suggesting that people try to ramp up their exercise Monday through Wednesday would be good for improving fitness.

```python
print('Monday: ', '\t', round(dayOfWeek.value_counts()[0]/dayOfWeek.count() * 100, 2), '\t', 'percent')
print('Tuesday: ', '\t', round(dayOfWeek.value_counts()[1]/dayOfWeek.count() * 100, 2), '\t', 'percent')
print('Wednesday: ', '\t', round(dayOfWeek.value_counts()[2]/dayOfWeek.count() * 100, 2), '\t', 'percent')
print('Thursday: ', '\t', round(dayOfWeek.value_counts()[3]/dayOfWeek.count() * 100, 2), '\t', 'percent')
print('Friday: ', '\t', round(dayOfWeek.value_counts()[4]/dayOfWeek.count() * 100, 2), '\t', 'percent')
print('Saturday: ', '\t', round(dayOfWeek.value_counts()[5]/dayOfWeek.count() * 100, 2), '\t', 'percent')
print('Sunday: ', '\t', round(dayOfWeek.value_counts()[6]/dayOfWeek.count() * 100, 2), '\t', 'percent')
```

```python
data = {'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
        'Percent': [
            round(dayOfWeek.value_counts()[0]/dayOfWeek.count() * 100, 2),
            round(dayOfWeek.value_counts()[1]/dayOfWeek.count() * 100, 2),
            round(dayOfWeek.value_counts()[2]/dayOfWeek.count() * 100, 2),
            round(dayOfWeek.value_counts()[3]/dayOfWeek.count() * 100, 2),
            round(dayOfWeek.value_counts()[4]/dayOfWeek.count() * 100, 2),
            round(dayOfWeek.value_counts()[5]/dayOfWeek.count() * 100, 2),
            round(dayOfWeek.value_counts()[6]/dayOfWeek.count() * 100, 2)
        ]}

df_active = pd.DataFrame(data)
```

#Heart rate breaking 126 beats per minute -- percentage by day of week
